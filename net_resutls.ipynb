```python
import cv2
from matplotlib import pyplot as plt
from core.services.lisam_net_service import LisamNetService
```

```python
ln_service = LisamNetService()
```

```python
img = cv2.imread("test_images/L.jpg", cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_inference = ln_service.inference_img(img)[0]
img_inference.to_dict()
```

```python
img_result = cv2.rectangle(
    img,
    (img_inference.box.xMin, img_inference.box.xMax),
    (img_inference.box.yMin, img_inference.box.yMax),
    color=(5, 63, 1), thickness=20
)
plt.imshow(cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB))
plt.show()
```
